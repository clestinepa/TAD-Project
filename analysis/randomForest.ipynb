```python
import pandas as pd
import psycopg2

param_dic = {
    "host"      : "localhost",
    "database"  : "postgres",
    "user"      : "postgres",
    "password"  : "bazepodataka"
}

con = psycopg2.connect(**param_dic)

sql_query = pd.read_sql('SELECT * FROM FLIGHTS_CLEANED', con)
df = pd.DataFrame(sql_query)
```

```python
#make the train test split
from sklearn.model_selection import train_test_split
X = df
y= df['delay_class']
del X['delay_class']
X_train, X_test, y_train, y_test = train_test_split(X,y ,
                                   random_state=21, 
                                   test_size=0.25, 
                                   shuffle=True)
print(X_train, y_train)
```

```python
#create a random forest, make predictions and display results 
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix

rf=RandomForestClassifier(max_depth = 13, n_estimators=300, random_state = 21)
rf.fit(X_train, y_train)

y_train_preds = rf.predict(X_train)
y_test_preds = rf.predict(X_test)

cm = confusion_matrix(y_test, y_test_preds)
print(classification_report(y_test,y_test_preds))
```
